```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Results folder:
ResultsFolder = "C:\\Users\\atpha\\Documents\\Postdocs\\Projects\\TES\Results\\Detroit\\Single\\"
```

```python
# Total costs:
m_salt = [20,40,60,80,100,120,140]
p_rating = [20,40,60,80,100,120,140,160,180,200]

cost_noTES = 3591.83499264179
li = []

for m in m_salt:
    for p in p_rating:
        cost_MgSO4 = pd.read_excel(ResultsFolder + 'Results_includeTES_True_' + str(p) + 'PR_' + str(m) + '_Building_id_5.xlsx')
        cost_MgSO4['saving'] = -(cost_MgSO4['total cost ($)'][0] - cost_noTES)
        cost_MgSO4 = cost_MgSO4['saving'][0]
        breakeven = cost_MgSO4/m*0.75*20
        df = pd.DataFrame()
        df = df.append({'salt mass' : int(m), 'power rating' : int(p), 'saving': cost_MgSO4, 'break even': breakeven}, ignore_index = True)
        li.append(df)

cost = pd.concat(li, axis=0)
cost.to_excel('C:\\Users\\atpha\\Documents\\Postdocs\\Projects\\TES\Results\\Detroit\\Compiled\cost2Ways.xlsx')
cost
```

```python

fig, ax = plt.subplots(figsize=(17,5))
#plt.scatter('salt mass', 'power rating', s='saving', c='lightsteelblue')
#cost.plot(kind='scatter', x='salt mass', y='power rating', c = 'saving', colormap='viridis')
cost.plot(kind='scatter', x='salt mass', y='power rating', c = 'saving', colormap='viridis')

ax.yaxis.set_visible(True)
ax.grid(False)
ax.set_xlim(0, 401)
plt.xlabel("Salt Mass (kg)",fontsize=20)
plt.ylabel("Cost Saving ($)", fontsize=20)
#ax.set_xticks(range(min(ordered_cost['building id']), max(ordered_cost['building id'])))
ax.tick_params(length=0, pad=8, labelsize=17)
ax.legend(fontsize=20, ncol=2,handleheight=0.5, labelspacing=0.15, loc='upper right')
fig.savefig('C:/Users/atpha/Documents/Postdocs/Projects/TES/Draft/images/cost_saving_2ways.png', dpi=300, bbox_inches='tight')
```
