```python
city = ['Detroit', 'Atlanta', 'Los Angeles', 'New York City', 'Minneapolis', 'Orlando', 'Phoenix', 'Seattle', 'Boston', 'Boulder', 'Chicago', 'Dallas']

lon = [lon_detroit, lon_atlanta, lon_la, lon_ny, lon_minneapolis, lon_orlando, lon_phoenix, lon_seattle, lon_boston, lon_boulder, lon_chicago, lon_dallas]
lat = [lat_detroit, lat_atlanta, lat_la, lat_ny, lat_minneapolis, lat_orlando, lat_phoenix, lat_seattle, lat_boston, lat_boulder, lat_chicago, lat_dallas]
tes = [tes_discharge_detroit*664.3, tes_discharge_atlanta*576.7, tes_discharge_la*3610.0, tes_discharge_ny*8509.9, tes_discharge_minneapolis*448.9,
       tes_discharge_orlando*311.0, tes_discharge_phoenix*1507.7, tes_discharge_seattle*803.4, tes_discharge_boston*701.8,
       tes_discharge_boulder*113.1, tes_discharge_chicago*2964.5, tes_discharge_dallas*1354.6]

tes_pct = [tes_discharge_detroit_pct*100, tes_discharge_atlanta_pct*100, tes_discharge_la_pct*100, tes_discharge_ny_pct*100, tes_discharge_minneapolis_pct*100,
           tes_discharge_orlando_pct*100, tes_discharge_phoenix_pct*100, tes_discharge_seattle_pct*100, tes_discharge_boston_pct*100,
           tes_discharge_boulder_pct*100, tes_discharge_chicago_pct*100, tes_discharge_dallas_pct*100]

frame = np.vstack((city, lon, lat, tes, tes_pct))
frame = pd.DataFrame(frame)
frame = frame.transpose()
frame.rename({0: 'City', 1: 'lon', 2: 'lat', 3: 'TES', 4: 'TES %'},axis=1, inplace=True)
frame = frame.astype({'lon': float, 'lat': float, 'TES': float, 'TES %': float})
frame
```

```python
# Spatial join station data with states:
from shapely.geometry import Point
from geopandas import GeoDataFrame
geometry = [Point(xy) for xy in zip(frame['lon'], frame['lat'])]
gpdframe = GeoDataFrame(frame, crs='epsg:4269', geometry=geometry)
costsaving_map = gpd.sjoin(gpdframe, us_states_in_NA, how="inner", op='intersects')

# Cost saving map for all cities:
minx, miny, maxx, maxy = us_states_in_NA.total_bounds

fig, ax = plt.subplots()
ax.set_aspect('equal')
us_states_in_NA.plot(ax=ax, color='white', figsize=(10,4), edgecolors='black', linewidth=0.5)
costsaving_map.plot(ax=ax, kind="scatter", x="lon", y="lat", s=700, edgecolors='black', #label="Cost Saving due to TES (%)",
                    c="TES %", cmap="Oranges", colorbar=True, legend=False, alpha=1, figsize=(10,4))
plt.axis('off')
f = plt.gcf()
cax = f.get_axes()[1]
cax.tick_params(labelsize=14)
cax.set_ylabel("Annual Load Shifting due to TES (%)", fontsize=14)
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
plt.tight_layout()
fig.savefig('C:/Users/atpha/Documents/Postdocs/Projects/TES/Draft/images/map_tes_pct.png', dpi=300, bbox_inches='tight')
```
