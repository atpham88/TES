```python
import glob
import pandas as pd

# Results folder:
ResultsFolder = "C:\\Users\\atpha\\Documents\\Postdocs\\Projects\\TES\\Results\\Detroit\\All\\"
salt_type = 'K2CO3'

# Combine results for all buildings into one place:
# all_files_no_TES = glob.glob(ResultsFolder + "*False*.xlsx")
all_files_w_TES = glob.glob(ResultsFolder + "*K2CO3*.xlsx")

# Total costs:
sheetname = 'total cost'
li = []
li2 = []
#for filename in all_files_no_TES:
#    building = filename.split("Building_id_", 1)[1]
#    building =  building.split('.xlsx')[0]
#    df = pd.read_excel(filename, sheet_name=sheetname)
#    df = df.rename(index={0: building})
#    li.append(df)
for filename in all_files_w_TES:
    building = filename.split("Building_id_", 1)[1]
    building =  building.split('.xlsx')[0]
    df = pd.read_excel(filename, sheet_name=sheetname)
    df = df.rename(index={0: building})
    li2.append(df)

#cost_noTES = pd.concat(li, axis=0)
cost_wTES = pd.concat(li2, axis=0)

#cost_noTES.to_excel('/home/anph/projects/TES/Results/Detroit/Compiled/' + 'total cost all building - no TES.xlsx')
cost_wTES.to_excel('/home/anph/projects/TES/Results/Detroit/Compiled/' + 'total cost all building'+'-' + salt_type +'.xlsx')

# TES discharge to serve load:
sheetname = 'TES'
li = []
for filename in all_files_w_TES:
    building = filename.split("Building_id_", 1)[1]
    building =  building.split('.xlsx')[0]
    df = pd.read_excel(filename, sheet_name=sheetname)
    tes_discharge = df['TES discharge (kWh)'].sum()
    tes_soc = df['TES SOC (kWh)'].sum()
    df2 = pd.DataFrame()
    df2 = df2.append({'building id' : int(building), 'TES discharge' : tes_discharge, 'TES SOC' : tes_soc}, ignore_index = True)
    li.append(df2)

tes_discharge = pd.concat(li, axis=0)
tes_discharge.to_excel('/home/anph/projects/TES/Results/Detroit/Compiled/' + 'TES discharge'+'-' + salt_type +'.xlsx')

# Load:
sheetname = 'load'
li = []
for filename in all_files_w_TES:
    building = filename.split("Building_id_", 1)[1]
    building =  building.split('.xlsx')[0]
    df = pd.read_excel(filename, sheet_name=sheetname)
    load = df['Load (kWh)'].sum()
    df2 = pd.DataFrame()
    df2 = df2.append({'building id' : int(building), 'Load' : load}, ignore_index = True)
    li.append(df2)

load = pd.concat(li, axis=0)
load.to_excel('/home/anph/projects/TES/Results/Detroit/Compiled/' + 'load.xlsx')

# Purchased electricity:
sheetname = 'purchased electricity'
li = []
li2 = []
#for filename in all_files_no_TES:
#    building = filename.split("Building_id_", 1)[1]
#    building =  building.split('.xlsx')[0]
#    df = pd.read_excel(filename, sheet_name=sheetname)
#    df = df.rename(columns={'Unnamed: 1': 'purchased electricity'})
#    purchased_electricity = df['purchased electricity'].sum()
#    df2 = pd.DataFrame()
#    df2 = df2.append({'building id' : int(building), 'purchased electricity' : purchased_electricity}, ignore_index = True)
#    li.append(df2)
for filename in all_files_w_TES:
    building = filename.split("Building_id_", 1)[1]
    building =  building.split('.xlsx')[0]
    df = pd.read_excel(filename, sheet_name=sheetname)
    df = df.rename(columns={'Unnamed: 1': 'purchased electricity'})
    purchased_electricity = df['purchased electricity'].sum()
    df2 = pd.DataFrame()
    df2 = df2.append({'building id' : int(building), 'purchased electricity' : purchased_electricity}, ignore_index = True)
    li2.append(df2)

#pur_elec_noTES = pd.concat(li, axis=0)
pur_elec_wTES = pd.concat(li2, axis=0)
#pur_elec_noTES.to_excel('/home/anph/projects/TES/Results/Detroit/Compiled/' + 'purchased electricity - no TES.xlsx')
pur_elec_wTES.to_excel('/home/anph/projects/TES/Results/Detroit/Compiled/' + 'purchased electricity'+'-' + salt_type +'.xlsx')

# HP output to serve load:
sheetname = 'HP output'
li = []
li2 = []
#for filename in all_files_no_TES:
#    building = filename.split("Building_id_", 1)[1]
#    building =  building.split('.xlsx')[0]
#    df = pd.read_excel(filename, sheet_name=sheetname)
#    hp_load = df['HP output to load (kWh)'].sum()
#    df2 = pd.DataFrame()
#    df2 = df2.append({'building id' : int(building), 'HP to load' : hp_load}, ignore_index = True)
#    li.append(df2)
for filename in all_files_w_TES:
    building = filename.split("Building_id_", 1)[1]
    building =  building.split('.xlsx')[0]
    df = pd.read_excel(filename, sheet_name=sheetname)
    hp_load = df['HP output to load (kWh)'].sum()
    hp_tes = df['HP output to TES (kWh)'].sum()
    hp_tot = df['HP output total (kWh)'].sum()
    df2 = pd.DataFrame()
    df2 = df2.append({'building id' : int(building), 'HP to load' : hp_load, 'HP to TES' : hp_tes, 'HP total': hp_tot}, ignore_index = True)
    li2.append(df2)

#hp_output_noTES = pd.concat(li, axis=0)
hp_output_wTES = pd.concat(li2, axis=0)
#hp_output_noTES.to_excel('/home/anph/projects/TES/Results/Detroit/Compiled/' + 'hp_output - no TES.xlsx')
hp_output_wTES.to_excel('/home/anph/projects/TES/Results/Detroit/Compiled/' + 'hp_output'+'-' + salt_type +'.xlsx')
```
